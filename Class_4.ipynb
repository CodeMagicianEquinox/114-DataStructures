Example:
Given "Rafael" your function should return leafaR
Given "stars" your function should return "srats"
Given "cool" your function should return "looc"

Your function must receive a string as a parameter and return a string (not a list).


```python
def inverting_string(word):
    inverted = ""
    for index in range(len(word)):            # 0 to len(word)-1
        inverted = inverted + word[len(word)-1-index]
    return inverted

print(inverting_string("Rafael"))
```

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, value):
        self.items.append(value)

    def pop(self):
        return self.items.pop()

    def size(self):
        return len(self.items)

    def peek(self):
        return self.items[len(self.items) - 1]

    def is_empty(self):
        return self.items == []

    def __str__(self):
        return f"Stack({self.items})"
```

```python
st1 = Stack()
st1.push(10)
st1.push(11)
st1.push(23)
st1.push(50)
print(st1)

print(st1.peek()) #Returns the last element in the stack
print(st1.size()) #gives me how many are in the stack
print(st1.pop()) #pops off the last element in the stack (at the bottom)
print(st1.peek()) #Returns the new last element in the stack
```

```python
def inverting_strings_using_stacks(word):
    inverted = ""
    st1 = Stack()

    for item in word:
        st1.push(item)

    while not st1.is_empty(): #stack is not empty. While loop continues
        inverted += st1.pop()
    
    return inverted

print (inverting_strings_using_stacks("Sky"))
```
